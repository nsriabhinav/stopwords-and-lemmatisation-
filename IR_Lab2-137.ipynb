### Pre-Processing

```python
from nltk import word_tokenize

word1 = word_tokenize(doc1)
word2 = word_tokenize(doc2)
print(word1[:5], word2[:5])
```

```python
no_punc_1 = []
for word in word1:
    temp_word = re.sub(r'[^\w\s]', '', word)
    if temp_word != '':
        no_punc_1.append(temp_word)
print(no_punc_1[:10])

no_punc_2 = []
for word in word2:
    temp_word = re.sub(r'[^\w\s]', '', word)
    if temp_word != '':
        no_punc_2.append(temp_word)
print(no_punc_2[:10])
```

```python
clean1 = list()
for word in no_punc_1:
    clean1.append(word.lower())

clean2 = list()
for word in no_punc_2:
    clean2.append(word.lower())

print(clean1[:10], clean2[:10])
```

```python
combined = [clean1,clean2]
combined
```

### Positional Index

```python
posn_ind = dict()

for curr, doc in enumerate(combined):
    for ind, word in enumerate(doc):
        #print(word, posn_ind)
        if word in posn_ind.keys():
            if (curr+1) not in posn_ind[word].keys():
                posn_ind[word][curr+1] = [ind]
            else:
                posn_ind[word][curr+1].append(ind)
        else:
            posn_ind[word] = dict()
            posn_ind[word][curr+1] = [ind]

posn_ind
```

### Edit Distance

```python
def levenshtein(s1, s2):
    dp = [[0]*(len(s2)+1) for _ in range(len(s1)+1)]
    for i in range(len(s1)+1):
        dp[i][0] = i
    for j in range(len(s2)+1):
        dp[0][j] = j
    
    for i in range(1,len(s1)+1):
        for j in range(1,len(s2)+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = min(dp[i-1][j]+1, dp[i][j-1]+1, dp[i-1][j-1])
            else:
                dp[i][j] = min(dp[i-1][j]+1, dp[i][j-1]+1, dp[i-1][j-1]+1)
    return(dp[-1][-1])

s1 = 'mice'
s2 = 'mouse'
print('The edit distance between "', s1 , '" and "', s2,  '" is: ', levenshtein('mice','mouse'))
```
