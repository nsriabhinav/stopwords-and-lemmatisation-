###**Vector space-based IR System** 

```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from tqdm import tqdm
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
df = pd.read_csv('sentences.csv')
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df = df.iloc[:1000]
```

```python
df.head()
```

```python
df.shape
```

```python
list_of_sentences = df['sentence'].tolist()
```

```python
punctuated = [] 

for sentence in list_of_sentences:
    tokens = word_tokenize(sentence)
    words = [word for word in tokens if word.isalnum()]
    punctuated.append(words)

punctuated
```

```python
lowercased = []

for sentence in punctuated:
    each_sentence = []
    for word in sentence:
        word = word.lower()
        each_sentence.append(word)
    lowercased.append(each_sentence)
```

```python
lowercased
```

```python
stop_words = stopwords.words('english')

processed = []

for sentence in lowercased:
    words = [w for w in sentence if not w in stop_words]
    processed.append(words)

processed
```

```python
l = list(range(1,1001))
```

```python
unique_words = []

for i in processed:
    for j in i:
        if j not in unique_words:
            unique_words.append(j)

unique_words
```

```python
data = {}

for i in range(1,1001):
    key = str(i)
    lst = [0] * 3120
    data[key] = lst
```

```python
df = pd.DataFrame(data, index = unique_words)
df
```

```python
for i in unique_words:
    for j in range(1000):
        c = 0
        if i in processed[j]:
            for k in range(len(processed[j])):
                if processed[j][k] == i:
                    c = c + 1
        df.loc[i,str(j+1)] = c
```

```python
doc_freq = []
for i in df.sum(axis = 1):
    doc_freq.append(i)
```

### **QUERY PROCESSING USING tf-idf**

Let the query for analysis be "accorded bottle abhi buried"




```python
query = ["accorded","bottle","abhi","buried"]

q_data = [0] *  df.shape[0]
query_df = pd.DataFrame(q_data,index = unique_words)
query_df
```

```python
query_df
```

```python
for i in unique_words:
    c = 0
    for j in range(len(query)):
        if i == query[j]:
            c = c + 1
    query_df.loc[i] = c
    
query_df
```

```python
query_doc_freq = []
for i in query_df.sum(axis = 1):
    query_doc_freq.append(i)

query_idf = []
for i in query_doc_freq:
    if i == 0:
        query_idf.append(0)
    else:
        query_idf.append(math.log10(1000/i))
        
query_idf
```

```python
import math

k = 0
for i in unique_words:
    if(query_df.loc[i] != 0).bool():
        m  = math.log(query_df.loc[i],10)
        query_df.loc[i] = query_idf[k] * (1+m)
    k = k + 1

query_df
```

```python
for i in range(1,1001):
    denom = math.sqrt(sum(df[str(i)]**2))
    for j in unique_words:
        df.loc[j,str(i)] = df.loc[j,str(i)]/denom
```

```python
query_denom = math.sqrt(sum(query_df[0]**2))
for j in unique_words:
    query_df.loc[j] = query_df.loc[j]/query_denom
```
